```python
import neo4j_connection

# Open and read the Cypher script from the file
cql_file_path = 'scripts/profiles_seed.cql'

with open(cql_file_path, 'r') as file:
    cypher_script = file.read()

# Display the content of the Cypher script (optional)
print(cypher_script)

# Split the script into individual queries
queries = [q.strip() for q in cypher_script.split(';') if q.strip()]

# Execute the Cypher script
for query in queries:
  result = neo4j_connection.execute_cypher_query(query)
  print(result)
```
